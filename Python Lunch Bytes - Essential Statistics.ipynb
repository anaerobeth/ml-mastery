## Why do some schools in Tennessee perform below average academically?

### Dataset
Variables: school_ratings and various demographic, school faculty, and income variables for every public school in middle Tennessee

```python
import pandas as pd
df = pd.read_csv('middle_tn_schools.csv')
df.describe()
```

### Descriptive and Inferential Statistics


```python
df[['reduced_lunch', 'school_rating']]\
    .groupby(['school_rating'])\
    .describe()\
    .unstack() # transpose for readability
```

### Correlation

```python
df[['reduced_lunch', 'school_rating']].corr()
```

### Visualization

```python
%matplotlib inline 
import matplotlib.pyplot as plt
import seaborn as sns

fig = plt.figure(figsize=(14,8)) # figure size
ax = fig.add_subplot(111) # axis information
ax.set_ylabel('school_rating') # sets label for y-axis

# boxplot with only these two variables
_ = df[['reduced_lunch', 'school_rating']].boxplot(by='school_rating', figsize=(13,8), vert=False, sym='b.', ax=ax)
```

```python
plt.figure(figsize=(14,8)) # set the size of the graph
_ = sns.regplot(data=df, x='reduced_lunch', y='school_rating')
```

```python
# create tabular correlation matrix
corr = df.corr()
_, ax = plt.subplots(figsize=(13,10)) 

# graph correlation matrix
_ = sns.heatmap(corr, ax=ax,
xticklabels=corr.columns.values,
yticklabels=corr.columns.values)
```

The variable school_rating, ranging from 0 to 5 stars, represent student performance data based on test scores.

Reduced_lunch is an indicator variable measuring the average percentage of students per school enrolled in a federal program that provides lunches for students from lower-income households. It can be used a a proxy for household income.

School_rating and reduced_lunch are negatively correlated.
